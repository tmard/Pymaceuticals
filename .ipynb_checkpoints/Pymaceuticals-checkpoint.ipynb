## Analysis

* Of the top four drugs that were analyzed in the dataset, Capomulin and Ramicane both appeared to be most effective at reducing tumor volumes. As well, in taking a look at the statistical data, there is less variance from the mean of the tumor volumes of Capomulin and Ramicane in comparison to the other drugs. This is further depicted when analyzing the data of the interquartile ranges and boxplots produced. 

* For Capomulin, there is a strong positive correlation between the moust weight and average tumor volume, indicating that as weight goes up, the average tumor volume goes up. Specifically for Mouse l509 treated with Capomulin, we see an initial slight increase in tumor volume, however, there is a decline in tumor volume as time goes on. 

* Evidence can presume that the drugs would be effective on both males and females, as the distribution was nearly 50/50. 

* Further considerations for ongoing studies and/or trials: Evaluation and analysis over longer treatment timepoints of Capomulin and Ramicane and analyzing if the data obtained is demonstrating statistical significance of tumor volumes, effects of diet and/or exercise on mice in conjunction with the drug, upon ceasing drug administration do tumor volumes change and hormonal effects on drug regimens (i.e., are there potential discrepancies between males and females based off of hormonal regulation).

## Prepare the Data

```python
# Import and dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
from scipy.stats import linregress
from sklearn import datasets
```

```python
%matplotlib inline
```

```python
# Read CSV files
mousefile = "Resources/Mouse_metadata.csv"
studyfile = "Resources/Study_results.csv"

mouse_df = pd.read_csv(mousefile)
study_df = pd.read_csv(studyfile)
```

```python
# Merge CSV file into single DataFrame
py_df = pd.merge(mouse_df,study_df, on=["Mouse ID","Mouse ID"], how="left")
py_df
```

```python
# Check for unique mice IDs and show number of unique
mice_id = py_df["Mouse ID"].nunique()
mice_id
```

```python
py_df.count()
```

```python
# Check mouse ID for duplicate time points
duplicate_row_df = py_df[py_df.duplicated(['Mouse ID', 'Timepoint'],keep=False)]
duplicate_row_df
```

```python
# Find the duplicated Mouse ID
duplicated_mouse_ids = duplicate_row_df['Mouse ID'].unique()
duplicated_mouse_ids
```

```python
# Display the duplicates
duplicated_mouse_row_s = py_df["Mouse ID"].isin(duplicated_mouse_ids)
py_df[duplicated_mouse_row_s]
```

```python
# Reveals total amount of duplicates
py_df.duplicated(['Mouse ID', 'Timepoint']).sum()
```

```python
# Cleaned DataFrame - remove the duplicates by taking out the duplicated rows 
clean_pydf = py_df[~duplicated_mouse_row_s]
clean_pydf
```

```python
# Display updated number of unique mice IDs
clean_miceid = clean_pydf["Mouse ID"].nunique()
clean_miceid
```

# Generate Summary Statistics

```python
# Generate summary statistics table - mean, median, variance, standard deviation and SEM of tumor volume for each drug

# Can use describe function to pull data - know describe returns mean (mean), mean (50%), standard deviation (std)
stats_drug = clean_pydf.groupby(['Drug Regimen'])['Tumor Volume (mm3)'].describe()

# Calculate variance
var_drug = clean_pydf.groupby(['Drug Regimen'])['Tumor Volume (mm3)'].var()

# Calculate SEM 
sem_drug = clean_pydf.groupby(['Drug Regimen'])['Tumor Volume (mm3)'].sem()

# Create Statistical DataFrame
stats = {'Mean Tumor Volume':stats_drug['mean'], 'Median Tumor Volume':stats_drug['50%'], 'Tumor Volume Variance':var_drug, 'Tumor Volume Std. Dev.': stats_drug['std'], 'Tumor Volume Std. Err.':sem_drug}
stats_df = pd.DataFrame(stats)
stats_df


# ** OF NOTE - could have used below, however, longer to compute - provides same information as using describe function
    # and pulling appropriate column from describe
# # Calculate mean
# mean_drug = clean_pydf.groupby(['Drug Regimen'])['Tumor Volume (mm3)'].mean()

# # Calculate median
# med_drug = clean_pydf.groupby(['Drug Regimen'])['Tumor Volume (mm3)'].median()

# # Calculate standard dev
# std_drug = clean_pydf.groupby(['Drug Regimen'])['Tumor Volume (mm3)'].std()
```

```python
# Generate summary statistics table - mean, median, variance, standard deviation and SEM of tumor volume for each drug

# Use aggregation method to produce same summary statistics in a single line
stats_df_agg = clean_pydf.groupby(['Drug Regimen'])['Tumor Volume (mm3)'].agg(['mean', 'median', 'var', 'std', 'sem'])

stats_df_agg.style.set_caption('Tumor Volume (mm3)')
```

# Quartiles, Outliers and Box Plot

```python
# Calculate final tumor volume of each mouse across - Capomulin, Ramicane, Infubinol, Ceftamin

# Find last timepoint (greatest time = max time) for each mouse ID & create dataframe
maxtime = clean_pydf.groupby(['Mouse ID'])['Timepoint'].max()
maxtime_df = pd.DataFrame(maxtime)

# Merge max dataframe on clean dataframe - place timepoint ON mouse ID to have ONLY max timepoints 
last_time_df = pd.merge(clean_pydf, maxtime_df, on=('Mouse ID', 'Timepoint'), how='right')
last_time_df 
```

```python
# Calculate the IQR and quantitatively determine if there are any potential outliers. 

# Put treatments into a list for for loop (and later for plot labels)
drug = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)
drug_values = []

# Locate each drug from our list 'drugs' in the orginal dataframe
for drug in drug:
    
    temp_df = last_time_df.loc[last_time_df['Drug Regimen'] == drug]

    # Create subset dataframes that narrow down to rows only containing the final timepoint for each regimen
    final_volume_df = temp_df.loc[temp_df['Timepoint'] == temp_df['Timepoint']]
    
    # Create a series that consists of all final volume values for each regimen
    values = final_volume_df['Tumor Volume (mm3)']
    drug_values.append(values)
    
    # Calculate and print the interquartile range (IQR) for each regimen
    quartiles = values.quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq
    print(f'IQR for {drug}: {iqr}.')
    
    # Find upper and lower bounds to help identify outliers for each regimen
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    print(f'Lower Bound for {drug}: {lower_bound}.')
    print(f'Upper Bound for {drug}: {upper_bound}.')
    
    # Quantitatively check for outliers
    outliers_count = (values.loc[(final_volume_df['Tumor Volume (mm3)'] >= upper_bound) | 
                                        (final_volume_df['Tumor Volume (mm3)'] <= lower_bound)]).count()
    print(f'Number of outliers for {drug}: {outliers_count}.')
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.

# X-axis is list of top drug regimen, y-axis is the drug_values obtained from above
drug = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']
vols = drug_values


# Edit characteristic of outlier
flierprops = dict(marker='o', markerfacecolor='r', markersize=8, markeredgecolor='black')


# Create boxplot and label
plt.boxplot(vols, flierprops=flierprops)

plt.title('Final Tumor Volumes for Top Drug Regimens')
plt.xticks([1,2,3,4], drug)
plt.xlabel('Drug Regimen')
plt.ylabel('Final Tumor Volume (mm3)')

plt.show()
```

# Scatter Plot

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen

# Find cases of only Capomulin
cap = clean_pydf.loc[clean_pydf["Drug Regimen"] == "Capomulin"]

# use groupby() and mean() to create a dataframe
cap_df = cap.groupby(['Mouse ID']).mean()

# Define x_axis by weight of each mouse in Capomulin
x_cap = cap_df['Weight (g)']

# Define y_axis by tumor volume of each mouse in Capomulin
y_cap = cap_df['Tumor Volume (mm3)']


# Create scatter plot and labels
plt.scatter(x_cap, y_cap)
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')

plt.show()
```

# Correlation Coefficient and Linear Regression

```python
# Calculate the correlation coefficient and linear regression model 

# Same code from scatter plot above:
cap = clean_pydf.loc[clean_pydf["Drug Regimen"] == "Capomulin"]
cap_df = cap.groupby(['Mouse ID']).mean()
x_cap = cap_df['Weight (g)']
y_cap = cap_df['Tumor Volume (mm3)']


# Calculate correction coefficeint between mouse weight and average tumor volume
correlation = st.pearsonr(x_cap,y_cap)
print(f"The correlation between mouse weight and the average tumor volume is {round(correlation[0],2)}.")


# Calculate linear regression 
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_cap, y_cap)
regress_values = x_cap * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))


# Create scatter plot and labels
plt.scatter(x_cap, y_cap)
plt.plot(x_cap,regress_values,"r-")
plt.annotate(line_eq,(20,35),fontsize=15,color="red")

plt.title('Average Tumor Volume versus Mouse Weight')
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')

plt.show()
```
